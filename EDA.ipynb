```python
diamonds = df.drop('Unnamed: 0', axis = 'columns')
diamonds.head()
```

```python
def print_unique_col_values(df):
       for column in df:
            print(f'{column}: {df[column].unique()}')

print_unique_col_values(diamonds)
```

```python
categorical = diamonds.drop(['carat_weight','depth_percent','table_percent','meas_length','meas_width','meas_depth'],axis = 'columns')
categorical.shape
```

```python
numerical = df[['carat_weight','depth_percent','table_percent','meas_length','meas_width','meas_depth','total_sales_price']]
numerical.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
list = ['carat_weight','depth_percent','table_percent','meas_length','meas_width','meas_depth']

fig, ax=plt.subplots(2,3,figsize=(12,4))
sns.set_theme(style="whitegrid")
count=0
for axid, col in enumerate(list):
    
    sns.boxplot(x=numerical[col],ax = ax[count,axid%3],color='blue')
    if axid==2:
        count = count+1
```

```python
numerical.describe()
```

```python
list=[]
for column in categorical.columns:
    list.append(column)

list.remove('total_sales_price')
list
```

```python
fig, ax = plt.subplots(6, 3, figsize=(15, 30))
count=0

for axid, col in enumerate(list):
    col_dist = diamonds[col].value_counts()
    piec_colors = sns.color_palette('pastel')[0:len(col_dist)]
    col_dist.plot(kind = 'pie', colors=piec_colors, autopct = '%1.2f%%', ax=ax[count,axid%3],label=col)
    if axid%3==2:
        count = count+1
```

```python
sns.heatmap(numerical.corr())
```

```python
fig, ax=plt.subplots(2,3,figsize=(12,10))
sns.set_theme(style="whitegrid")
count=0
for axid, col in enumerate(numerical.columns):
    
    sns.distplot(numerical[col],ax = ax[count,axid%3])
    if axid==2:
        count = count+1
```

```python
X = diamonds.drop('total_sales_price',axis='columns')
fig, ax=plt.subplots(5,5,figsize=(45,30))
sns.set_theme(style="whitegrid")
count=0
for axid, col in enumerate(X.columns):
    
    sns.scatterplot(x=X[col], y=diamonds['total_sales_price'],ax = ax[count,axid%5])
    if axid%5==4:
        count = count+1
```
