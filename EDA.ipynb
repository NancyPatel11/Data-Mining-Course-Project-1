```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('diamonds.csv',index_col=0)
pd.set_option('display.max_columns',None)
df.head()
```

```python
# Function to print all the unique values for categorical columns
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}')
```

```python
print_unique_col_values(df)
```

```python
# Making a list of all categorical columns
obj_columns=[]
for column in df:
    if df[column].dtypes=='object':
        obj_columns.append(str(column))
```

```python
# Plotting countplot for all categorical columns
fig, ax = plt.subplots(6, 3, figsize=(15, 30))
ax=ax.ravel()

for axid, col in enumerate(obj_columns):
    sns.countplot(x=col,data=df, ax=ax[axid])
    plt.title=col
```

```python
# Plotting distribution on all categorical columns in the form of a pie chart
fig, ax = plt.subplots(6, 3, figsize=(15, 30))
ax=ax.ravel()

for axid, col in enumerate(obj_columns):
    col_dist = df[col].value_counts()
    piec_colors = sns.color_palette('pastel')[0:len(col_dist)]
    col_dist.plot(kind = 'pie', colors=piec_colors, autopct = '%1.2f%%', ax=ax[axid])
    plt.title=col
```

From the above analysis we can conclude that columns 
1) fluor_color
2) fancy_color_dominant_color
3) fancy_color_secondary_color
4) fancy_color_overtone
5) fancy_color_intensity
6) eye_clean
7) culet_condition

have very high proportion of unknown values.

Apart from that we can see that the columns like
1) lab
2) cut
3) symmetry

are significantly imbalabced.

```python
# Function to list all numerical columns
num_columns=[]
for col in df:
    if df[col].dtypes=='float64' or df[col].dtypes=='int64':
        num_columns.append(col)
```

```python
df.describe()
```

```python
# Plotting histgram for each numerical columns to get a better idea about the distribution of the numeric data
fig, ax = plt.subplots(3, 3, figsize=(15, 30))
ax=ax.ravel()

for axid, col in enumerate(num_columns):
    sns.distplot(df[col], ax=ax[axid])
    plt.title=col
```

```python
print('Skewness for each numerical column')
for col in num_columns:
    print(col+' : '+str(df[col].skew()))
```

Columns
1) meas_depth
2) total_sales_price

seems to have highly right skewed data distribution.

```python
# Ploting box plot for each numerical column
fig, ax = plt.subplots(3, 3, figsize=(15, 30))
ax=ax.ravel()

for axid, col in enumerate(num_columns):
    sns.boxplot(df[col], ax=ax[axid])
    plt.title=col
```

```python
upper=[2.035,66.95,64.5,9.35,7.785,5.055,-5415.5]
lower=[-0.725,57.75,52.5,1.35,2.225,1.255,11580.5]

print('Outlier in percentage')
for i in range(7):
    count=0
    for val in df[num_columns[i]]:
        if val>upper[i] or val<lower[i]:
            count=count+1

    print(num_columns[i]+' : '+str(count/len(df[num_columns[i]])*100)+' %')
```

We can observe that there are high number of outliers in the column
1) total_sales_price
